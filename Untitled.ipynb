```python
#from pycaret.datasets import get_data
from pycaret.regression import *
import pandas as pd
```

```python
d = get_data('diamond')
```

```python
data = pd.read_csv('diamonds.csv')
data.head()
```

```python
data.loc[data['clarity'] == 'I1', 'clarity'] = 1
data.loc[data['clarity'] == 'SI2', 'clarity'] = 2
data.loc[data['clarity'] == 'SI1', 'clarity'] = 3
data.loc[data['clarity'] == 'VS2', 'clarity'] = 4
data.loc[data['clarity'] == 'VS1', 'clarity'] = 5
data.loc[data['clarity'] == 'VVS2', 'clarity'] = 6
data.loc[data['clarity'] == 'VVS1', 'clarity'] = 7
data.loc[data['clarity'] == 'IF', 'clarity'] = 8

def change_color(x):
    return 74 - ord(x)
data['color'] = data.apply(lambda row : change_color(row['color']),axis=1)

data.drop(['Unnamed: 0'],axis=1,inplace=True)
data.head()
```

```python
s = setup(data,'price',session_id=42)
```

```python
best = compare_models()
```

```python
# check the residuals of trained model
plot_model(best, plot = 'residuals_interactive')
```

```python
# check feature importance
plot_model(best, plot = 'feature')
```

```python
# finalize the model
final_best = finalize_model(best)
# save model to disk
save_model(final_best, 'diamond-pipeline')
```
